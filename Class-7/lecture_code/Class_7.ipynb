```python
#Linear Regression
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

#Load the dataset
df = sns.load_dataset('tips')
print(df)
X = df['total_bill'].values.reshape(-1, 1)
print(X)
y = df['tip'].values
print(y)

#Build the relationship between X and Y
model = LinearRegression()
model.fit(X, y)

#Predict the values
residuals = model.predict(X) - y
sm.qqplot(residuals, line='s')
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Sample data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 6, 8])

#Fit model
model = LinearRegression()
model.fit(X.reshape(-1, 1), y)
#plot
plt.scatter(X,y, color='blue', label='Data')
plt.plot(X,model.predict(X.reshape(-1, 1)), color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()



```

```python
#Build the Linear Regression Model to predict houses prices
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Define Data set
square_feet = np.array([1000, 1500, 2000, 2500, 3000])
house_prices = np.array([200000, 250000, 300000, 350000, 400000])

#Build the model / Train the model
model = LinearRegression() #Initilaize the model
model.fit(square_feet.reshape(-1,1), house_prices) #Train the model

#Make a prediction
predicted_price = model.predict(np.array([2000]).reshape(-1,1))
print(f'Predicted House price for 2000 sqft: ${predicted_price}')
```

```python
plt.scatter(square_feet, house_prices, color='blue', label='Actual Data')
plt.plot(square_feet, model.predict(square_feet.reshape(-1,1)), color='red', label='Regression Line')
plt.xlabel('Square Feet')
plt.ylabel('House Prices')
plt.title('House Price Prediction')
plt.legend()
plt.show()
```

##Task-1

#### 1. Import NumPy library.

#### 3. Define a function named linear_function with parameters x, m, and c
#### 3.1 Return the result of the linear equation y = mx + c

```python
import numpy as np
import matplotlib.pyplot as plt

def linear_function(x, c=3, m=2):
  return (m*x + (c))
```

#Task-2

### 1. Generate 100 evenly spaced values between -10 and 10 using NumPy
### 2. Compute corresponding y values using the linear_function



```python
x_values = np.linspace(-10, 10, 100)
y_values = linear_function(x_values)
```

#Task-3

### 1. Plot the linear function with a label
### 2. Label the x-axis as "x"
### 3. Label the y-axis as "y"
### 4. Display a legend for the plotted function
### 5. Enable grid lines on the plot
### 6. Display the plot


```python
plt.plot(x_values, y_values, label="y = 2x + 3", color='b', linestyle='-', linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Function")
plt.legend()
plt.grid()
plt.show()
```
